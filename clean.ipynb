```python
forms = ['AVN_Model_SpecialRequestForm',
        'AVN_Model_BookingCar',
        'AVN_Model_OnlineBooking']
```

```python
# лиды на AVN без группировки в array
# ? timestamp  as ts,

avn_leads_req = f"""
SELECT domainWithoutWWW(sess_refhost) as domain,  
            toISOWeek(date) as week,
            
            uniqIf((uid,timestamp), (data_category in ({forms}) and positionCaseInsensitive(data_action, 'success')>0
            or data_category='AVN_Model_ShowPhone')) as total_leads,
            
            uniqIf(uid, data_category in ({forms}) and positionCaseInsensitive(data_action, 'success')>0 
            or data_category='AVN_Model_ShowPhone') as unique_leads,
            
            toDateTime(sess_start) as sess_start,
            toStartOfWeek(date) as date,
            if(uaparser_is_pc==1, 'PC','Mobile') as device,
            data_category,
            sess_type,
            
            uid
            
FROM gaevents 
        
WHERE domainWithoutWWW(page_url)== 'cars.skoda-avto.ru'
            AND extractURLParameter(page_url,'dealerid')!=''
            AND projectId = 598135432
            AND date >= '2021-09-01'
            AND data_category in ({forms}) OR data_category='AVN_Model_ShowPhone'
            
    GROUP BY week, date, domain, sess_type, data_category, device, uid, sess_start
"""
```

```python
avn_leads = client.query_dataframe(avn_leads_req)
```
